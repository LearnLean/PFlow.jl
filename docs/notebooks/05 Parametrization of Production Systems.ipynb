# Parametrization of Production Systems

*Paul Bayer, v0.1, 2017-08-24*

Since Orlicky's work<sup>1</sup> in the 1960's and the subsequent rise of [Materials requirements planning – MRP](https://en.wikipedia.org/wiki/Material_requirements_planning) production systems are managed, planned and controlled by lists of

- production plans,
- schedules and
- inventories.

Now the **basic idea** of `PFlow` is simple: if it is possible to manage and to control real production systems by such lists, it must be possible to reduce the input to their simulation to the same information. The simulation should run on (a subset of) the same information, which is needed to fed and to control the real system. In order to deliver correct results, the simulation must be able:

1. to take the same information as fed into the real system,
2. represent the inner workings of the real system and
3. generate a similar output as the real system, which
4. can be analyzed in order to assess the system performance.

All this has to be done in "no time" in order to support the decisions which have to be done in managing the real system.


## Product manufacturing flow

For manufacturing systems, the situation is described above and usually enough information is available in order to describe the inputs, the structure and the inner workings of the planned or given production system.


## Product development flow

Astonishingly – even if usually seen as another world – the situation for projects is not so different. In practice 

- there is not so much integrated planning as in manufacturing,
- variation in execution times is bigger and
- multitasking and concurrent usage of resources by projects are common.

But the inputs and basic information are similar lists of

- projects,
- project plans,
- resources.

One major difference is the longer timeline, during which new projects arrive, get fed into the system and have to be managed through. There is no master production schedule (MPS) like in production.


## Service flow

If it is possible to simulate a multiproject system with this approach, it should be possible also to extend it to the simulation of service systems, health care (eg. hospitals, medical practices) and the like.


## Usability and convenience

With `PFlow's` approach comes a major advantage: usability and convenience. Without having to care for the inner workings of a system, a user can simply define its components, the use cases (eg. orders, customer types), the demand, the policies, fed it into PFlow, start the simulation and look what happens.
